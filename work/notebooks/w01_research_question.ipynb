## 2. The question: decision, action, cost of a wrong call

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Show the scale of the problem to justify why a human reviewer needs our help
import pandas as pd

# Load the data here first
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Show the scale of the problem to justify why a human reviewer needs our help
total_pages = len(df)
top_20_pages = df[df['avg_position'] <= 20]

print(f"Total pages in the dataset: {total_pages:,}")
print(f"Pages ranking in the top 20: {len(top_20_pages):,}")
print("Conclusion: A human reviewer cannot manually check tens of thousands of pages. Our ranking model is necessary to prioritize the workflow.")
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

My code shows that for the worst 10% of the top pages, their CTR are siginificantly worse the median level websites, which means there might exist critical fixes that can improve it CTR

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Filter for legitimate top-3 ranking pages with significant traffic volume
top_ranking = df[(df['avg_position'] > 0) &
                 (df['avg_position'] <= 3) &
                 (df['impressions_90d'] > 1000)]

# Calculate the percentiles for Click-Through Rate
ctr_10 = top_ranking['ctr'].quantile(0.10)
ctr_50 = top_ranking['ctr'].median()
ctr_90 = top_ranking['ctr'].quantile(0.90)

print(f"Total Top 3 Pages (>1k impressions): {len(top_ranking)}")
print(f"10th Percentile CTR: {ctr_10:.1%}")
print(f"50th Percentile CTR (Median): {ctr_50:.1%}")
print(f"90th Percentile CTR: {ctr_90:.1%}")
```
